```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
import statsmodels.api as sm
from statsmodels.stats.multicomp import pairwise_tukeyhsd


file_path=r"C:\Users\petro\Desktop\Applied Statistics Project\europe_covid.csv"
df=pd.read_csv(file_path)

gg={'Anatoliki Eurwph': ['Albania', 'Bulgaria', 'Romania', 'Serbia', 'Ukraine'],
    'Dytiki Eurwph': ['Germany', 'France', 'Belgium', 'Netherlands'],
    'Northern Europe': ['Norway', 'Finland', 'Sweden', 'Denmark'],
    'Notia Eurwph': ['Italy', 'Spain', 'Portugal', 'Greece'],
    'Kentriki Eurwph': ['North Macedonia', 'Montenegro']}
region_map={}
for region, countries in gg.items():
    for country in countries:
        region_map[country]=region

df["Region"]=df["Country/Other"].map(region_map)
df_region = df.dropna(subset=["Region"])


df_region["Mortality_Rate"]=df_region["Total Deaths"]/df_region["Total Cases"]
df_region["Recovery_Rate"]=df_region["Total Recovered"]/df_region["Total Cases"]
df_region["Cases_per_100k"]=df_region["Total Cases"]/df_region["Population"]*100000

mean_mortality=df_region["Mortality_Rate"].mean()
print("Μέσο ποσοστό θνητότητας COVID-19 στην Ευρώπη: ",mean_mortality)

model=ols("Cases_per_100k ~ C(Region)", data=df_region).fit()
anova_table=sm.stats.anova_lm(model, typ=2)
print("\nΑΠΟΤΕΛΕΣΜΑΤΑ ANOVA:\n")
print(anova_table)

alpha=0.05
p_value=anova_table["PR(>F)"][0]
if p_value<alpha:
    print("\n Υπάρχουν στατιστικά σημαντικές διαφορές (p < 0.05) μεταξύ γεωγραφικών περιοχών.")
else:
    print("\n Δεν υπάρχουν στατιστικά σημαντικές διαφορές (p ≥ 0.05) μεταξύ γεωγραφικών περιοχών.")

plt.figure(figsize=(10,6))
sns.boxplot(x='Region', y='Cases_per_100k', data=df_region, palette='Set2')
plt.title("Κατανομή Κρουσμάτων ανά 100.000 κατοίκους ανά Γεωγραφική Περιοχή")
plt.ylabel("Κρούσματα ανά 100.000")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(r"C:\Users\petro\Desktop\cases_per_region_boxplot.png",dpi=300,bbox_inches='tight')
plt.show()
```
